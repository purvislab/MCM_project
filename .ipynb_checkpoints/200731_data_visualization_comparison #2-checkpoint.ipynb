# 1. normalize the value

```python
dataHP1BFinal=dataHP1B.loc[((dataHP1B.day!='d20190625') & (dataHP1B.day!='d201907012') 
                          & (dataHP1B.day!='d20190716') & ['G1' in x for x in dataHP1B.category]),:]
```

```python
dataHP1BFinal.day.unique()
```

```python
dataHP1BFinal.temp.describe()
```

```python
from sklearn import preprocessing
```

```python
dataHP1BFinal['temp']=(dataHP1BFinal.nuc_mcm_het/dataHP1BFinal.nuc_DAPI_het)/(dataHP1BFinal.nuc_mcm_total/dataHP1BFinal.nuc_DAPI_total)
```

```python
dataHP1BFinal['temp1']=(dataHP1BFinal.temp-dataHP1BFinal.temp.min()) / (dataHP1BFinal.temp.max() - dataHP1BFinal.temp.min())
```

```python
%matplotlib notebook
dataHP1BFinal['temp']=(dataHP1BFinal.nuc_mcm_het/dataHP1BFinal.nuc_DAPI_het)/(dataHP1BFinal.nuc_mcm_total/dataHP1BFinal.nuc_DAPI_total)
ax=sns.boxplot(x='category',y='temp1',data=dataHP1BFinal,hue='day',boxprops=dict(alpha=.5),palette='PuRd')
sns.swarmplot(x='category',y='temp1',data=dataHP1BFinal,hue='day',dodge=True,palette='PuRd')


handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:5], labels[:5])
plt.ylabel('MCM/DAPI [het/nuc]')
plt.xlabel('G1 phases')
plt.title('Relative MCM loading in heterochromatin to nucleus pixels')
```

## Anova test and hypothesis
- **Null hypotheses**: Groups means are equal (no variation in means of groups)
- **Alternative hypotheses**: At least, one group mean is different from other groups

```python
formula = 'temp1 ~ day + category'
model = sm.formula.ols(formula, data=dataHP1BFinal).fit()

# Perform ANOVA and print table
aov_table = sm.stats.anova_lm(model, typ=2)
print(aov_table)
```

The p-value of the ANOVA test shows us that there is significant evidence to reject the null hypothesis and conclude that the group mean is different from the other groups for both variables, day and category. 

### assumption tests hypothesis
#### test of normal distribution of residuals
The Shapiro-Wilk test can be used to check the normal distribution of residuals.
Null hypothesis: data is drawn from normal distribution.

```python
import scipy.stats as stats
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
```

#### test of homogeneity of variance
Null hypothesis: samples from populations have equal variances.

```python
w, pvalue = stats.bartlett(dataHP1BFinal.loc[dataHP1BFinal['category']=='earlyG1','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['category']=='middleG1','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['category']=='lateG1','temp1'])
print(w, pvalue)
```

```python
w, pvalue = stats.bartlett(dataHP1BFinal.loc[dataHP1BFinal['day']=='d20190506','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['day']=='d20190607','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['day']=='d20190609','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['day']=='d20200203','temp1'],
                           dataHP1BFinal.loc[dataHP1BFinal['day']=='d20191203','temp1'])
print(w, pvalue)
```

### conclusion of assumption tests
The p-value for normal distribution test show that there is significant evidence for us to reject the null hypothesis of data having normal distribution. This conclusion shows us that the data does not have normal distribution. 

Also, the test for equal variance in category and day show that there is significant evidence for us to reject the null hypothesis of data variables have an equal variances in their values. This conclusion means that there isn't any equal variance among the data values.

```python
dataHP1BFinal['temp2']=dataHP1BFinal['temp1']=(dataHP1BFinal.temp-randomDayMin) / (randomDayMax - randomDayMin)
```
