```python
myDF['cell id for tracking']=0
```

```python
for i, row in myDF.iterrows():
    try:
        myDF.at[i,'cell id for tracking'] = int(myDF.at[i,'original cell name'].split("1125-")[1].split("-")[1])
    except:
        myDF.at[i,'cell id for tracking'] = int(myDF.at[i,'original cell name'].split("1125-")[1].split("-")[1].split("(")[0])
```

```python
for i, row in myDF.iterrows():
    
    try:
        file=myDF.at[i,'movie']
        currFile=f'{myDir}\\{file}.csv'
        currDF=pd.read_csv(currFile)
        myDF.at[i,'molecular age 10']= currDF.at[myDF.at[i,'cell id for tracking']-1, 'MeanRatio']
    except:
        myDF.at[i,'movie']='191125pcnat_dhbr_cdc6vxy08'
        file=myDF.at[i,'movie']
        currFile=f'{myDir}\\{file}.csv'
        currDF=pd.read_csv(currFile)
        myDF.at[i,'molecular age 10']= currDF.at[myDF.at[i,'cell id for tracking']-1, 'MeanRatio']

```
