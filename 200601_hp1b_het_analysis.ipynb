```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from scipy import stats
```

```python
file=f'{bigDir}\cellinfo_200203_v7.csv'
```

```python
data=pd.read_csv(file)
```

# data visualizations 

```python
data.head()
```

```python
data.loc[(data.category=='middel G1'),'category']='middleG1'
```

```python
data.loc[(data.category=='earlr G1'),'category']='earlyG1'
```

```python
%matplotlib notebook
sn.scatterplot(x='nuc_hp1b_het',y='nuc_hp1b_total',data=data.loc[(['G1' in x for x in data.category]),:],hue='category')
plt.xlabel('HP1B Signal in Heterochromatin pixels')
plt.ylabel('HP1B Signal in nucleus pixels')
plt.title('HP1B Signal in het vs nucleus pixels')
```

```python
%matplotlib notebook
data['temp']=data.nuc_hp1b_het/data.nuc_hp1b_total
sn.boxplot(x='group',y='temp',data=data.loc[(['G1' in x for x in data.category]),:],hue='category')
plt.xlabel('Groups')
plt.ylabel('HP1B Signal in het pixels / nucleus pixels')
plt.title('HP1B Signal in het/nuc pixels in two groups')
```

```python
%matplotlib notebook
data['temp']=data.nuc_hp1b_het/data.nuc_hp1b_total
sn.boxplot(x='group',y='temp',data=data.loc[(['G1' in x for x in data.category]),:])
plt.xlabel('Groups')
plt.ylabel('HP1B Signal in het pixels / nucleus pixels')
plt.title('HP1B Signal in het/nuc pixels in two groups')
```

```python
%matplotlib notebook
data['temp']=data.nuc_hp1b_het/data.nuc_hp1b_total
sn.violinplot(x='group',y='temp',data=data.loc[(['G1' in x for x in data.category]),:])
plt.xlabel('Groups')
plt.ylabel('HP1B Signal in het pixels / nucleus pixels')
plt.title('HP1B Signal in het/nuc pixels in two groups')
```
