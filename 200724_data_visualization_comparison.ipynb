## One-way anova test 

### hypothesis
- Null hypotheses: Groups means are equal (no variation in means of groups)
- Alternative hypotheses: At least, one group mean is different from other groups

One way anova test is used to see if the mean of the groups are significantly different or not. Here, I will test for the three G1 phase group. We will reject the null hypothesis if the p-value is significant at 5% (less than 0.05)

```python
import scipy.stats as stats
fvalue, pvalue = stats.f_oneway(dataFinal1.loc[dataFinal1.category=='earlyG1','temp'], 
                                dataFinal1.loc[dataFinal1.category=='middleG1','temp'],
                                dataFinal1.loc[dataFinal1.category=='lateG1','temp'])
print(fvalue, pvalue)
```

## Tukey HSD Test

With this test, it will be seeing the results of 2 groups paired with each other and see the mean different in the two groups. Same as ANOVA test, if the p-value is significant at 5% (0.05), we will have enough evidence to reject the null hypothesis

## Test assumptions
One of the assumptions for the ANOVA test is the there is an equal variance between the groups that are being tested. 

Making a linear regression model for temp (proportion of MCM/DAPI) vs category in G1phase.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# reshape the d dataframe suitable for statsmodels package 
data_melt = pd.melt(dataFinal1.reset_index(drop=True), id_vars=['category'], value_vars=['temp'])
# replace column names
data_melt.columns = ['index','category','temp']
# Ordinary Least Squares (OLS) model
model = ols('temp ~ C(index)', data=data_melt).fit()
```

```python
data_melt.head()
```

### test of normal distribution of residuals
The Shapiro-Wilk test can be used to check the normal distribution of residuals. 
- Null hypothesis: data is drawn from normal distribution.


```python
import scipy.stats as stats
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
```

Since the p-value for this this is significant at 5% level, we do not have enough evidence to reject the null hypothesis and would have to conclude that the data is not drawn from normal distribution. However, for our case of interest, this conclusion is valid since the experiment days were different, and the signal values and the distribution would differ a lot by the days and the categories. 

### test of homogeneity of variance
- Null hypothesis: samples from populations have equal variances.

Although we know we don't have enough evidence to conclude that the data is drawn from normal distribution, we will still see if the variance amoung the populations is same. 

```python
w, pvalue = stats.bartlett(dataFinal1.loc[dataFinal1['category']=='earlyG1','temp'],
                           dataFinal1.loc[dataFinal1['category']=='middleG1','temp'],
                           dataFinal1.loc[dataFinal1['category']=='lateG1','temp'])
print(w, pvalue)
```

Like expected, the p-value is significant at 5%, leading us to reject the null hypothesis and conclude that the samples from the populations do not have equal variances. However, just like with the normal distributions, unequal variances are expected since each category of G1 phase is composed of different days of experiment, most likely having different distributions.
