## Creating array/list for cell ids and tracking file numbers

```python
tracking_nums = []
cell_ids = []
for i in range(len(cells)):
    if i == 70 or i == 90:
        cell_ids.append(1011)
    elif i == 173 or i == 190:
        cell_ids.append(1001)
    elif math.isnan(float(cells[i][0])):
        cell_ids.append(None)
    else:
        cell_ids.append(int(cells[i][0]))
    
    tracking_nums.append(cells[i][1])
```

## function retrieveData 
in: tracking, cell_id
out: one array of arrays of mean and mean ring

This is done by calling function giveData. This function mainly checks for valid input. 
ex) [[mean, mean ring]]

```python
def retrieveData (tracking, cell_id):
    
    if cell_id is None or math.isnan(tracking):
        return "CANNOT FIND THE DATA - LACKING TRACKING FILE AND/OR VALID CELL ID #"
    elif cell_id == 1001:
        cell_id = 0
        return giveData(tracking, cell_id)
    elif cell_id == 1011:
        cell_id = 10
        return giveData(tracking, cell_id)
    else:
        return giveData(tracking, cell_id)
```

## function giveData
in: tracking, cell_id
out: one array of arrays of mean and mean ring

This function returns the data back to the function **retrieveData**. This function is not called by itself in this program. 
It also checks if *trackingData* is a NoneObject or not. 
ex) [[mean, mean ring]]

```python
def giveData (tracking, cell_id):
    trackingData = openingFiles(tracking)
   

    if trackingData is None:
        return "CANNOT FIND THE DATA - LACKING TRACKING FILE AND/OR VALID CELL ID #"
    else: 
        df = pd.DataFrame(trackingData, columns = ['Area', 'Mean', 'StdDev','Mode', 'Slice', 'Cell ID #', 'Movie Slice', 'Mean ring', 'Median ring', 'Std ring'])

        data_mean = []
        
        for i in trackingData.index:
             if cell_id == df['Cell ID #'][i]:
                data_mean.append([df['Mean'][i], df['Mean ring'][i]])

        return data_mean
```
